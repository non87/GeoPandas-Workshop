```python
fl = "tl_2010_17031_bg10.shp"
blk = gpd.read_file(fl)
```

```python
print(blk.shape)
print(blk.head())
blk['GEOID10'] = blk['GEOID10'].apply(lambda x: "1500000US" + x)
```

```python
race_fl = 'DECENNIALSF12010.H6_data_with_overlays_2020-06-02T210055.csv'
race = pd.read_csv(race_fl)
print(race.shape)
#print(race.head())
new_names = {'GEO_ID':'GEOID10', "H006001": 'TOTAL', "H006002":'WHITE', "H006003":'BLACK'}
race = race.rename(columns = new_names)
race = race.loc[:, ['GEOID10', 'TOTAL', 'WHITE', 'BLACK']]
#print(race.head())
race = race.drop(0)
print(race.head())
print(race.dtypes)
to_convert = ['TOTAL', 'WHITE', 'BLACK']
for col in to_convert:
    race[col] = race[col].astype("int64")
print(race.dtypes)
```

```python
race['BLACK'].max()
FIPS = "170318435001"
FIPS = "1500000US" + FIPS
race.loc[race['GEOID10'] == FIPS, 'TOTAL'] = 11373
race.loc[race['GEOID10'] == FIPS, 'BLACK'] = 7603
race.loc[race['GEOID10'] == FIPS, 'WHITE'] = 3188
print(race.loc[race['GEOID10'] == FIPS, :])
```

```python
print(blk['GEOID10'])
print(race['GEOID10'])
blk = blk.merge(race, on = 'GEOID10')
```

```python
blk.plot(column = 'TOTAL')
blk.plot(column = 'BLACK')
blk.plot(column = 'WHITE')
```
